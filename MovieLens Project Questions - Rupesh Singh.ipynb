### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use `sum(axis=1)` store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.  

```python
df = df_movie
#adding release year
df_release_year = df_movie["release date"].str.split("-", n=2,expand = True)[2].reset_index()
df["release year"] = df_movie["release date"].str.split("-", n=2,expand = True)[2]

#getting sum of release year 
df_release_year = df.groupby('release year').sum()

#adding total column having count of movies per year
df_release_year['Total'] = df.groupby('release year').size()

#removing unnecessary columns
df_release_year.drop(['movie id','unknown'], axis=1,inplace=True)

#calculating percentage
df_release_year = df_release_year.div(df_release_year['Total'], axis= 0) * 100
df_release_year.drop(['Total'], axis=1,inplace=True)

#verify the data
df_release_year.head(200)
```

```python
#plotting the heatmap
plt.figure(figsize=(30,15))
sns.set(font_scale=2)
sns.heatmap(df_release_year.transpose(),cmap='coolwarm', cbar_kws={'label': 'Percent of Releases'},annot=False)
plt.show()
```

```python
#removing the unwanted columns
dfr = df_rating
dfr.drop(['user id','timestamp'], axis=1,inplace=True)
```

```python
#getting rating count and mean (average) for each movie 
dfr_mean = dfr.groupby('movie id').agg(['mean','count'])
dfr_mean.columns = ['average rating','count of rating']

#filtering the data having rating count > 100
dfr_mean = dfr_mean[dfr_mean['count of rating']>100]

#sorting the data based on mean and seleting top 25
dfr_mean = dfr_mean.sort_values(by=['average rating'], ascending=False).head(25)

#merge join with the movie data set
df_movie_with_title = pd.merge(dfr_mean, df_movie, on=["movie id"])

df_movie_with_title[['movie title','average rating', 'count of rating']]
```

### 10. See gender distribution across different genres check for the validity of the below statements - 10 marks

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women

**compare the percentages**

1. Merge all the datasets

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
#merge join with the movie data set
df = pd.merge(df_rating, df_movie, on=["movie id"])
df = df.drop(['movie id','release date','movie title','rating','timestamp'], axis=1)

#group by each user and get sum by genre
df2 = df.groupby('user id').sum()
df2.head()

#merge the data with user dataset to get gender
df4 = pd.merge(df2, df_user, on=["user id"])
df4 = df4.drop(['age','occupation','zip code'], axis=1)
df4.head()

#group the data by gender and get sum for genre
df5 = df4.groupby('gender').sum()
df5 = df5.reset_index()
df5.head()

#getting total rating from total male and female
user_ratings=pd.merge(df_rating,df_user,  how='inner', on=['user id'])
ratings_gender= user_ratings.groupby('gender').agg({'user id': ['count']}).reset_index()
ratings_gender.columns = ['gender','count']
ratings_gender.head()

#getting the final data frame
final_df = pd.merge(ratings_gender,df5, how='inner',on=['gender'] )
final_df.head()
```

```python
#analysing the statement - Men watch more drama than women

# Female: ~43%
# Male: ~39%

# Men watch more drama than women - FALSE
df_drama = final_df['Drama'].div(final_df['count'])
df_drama.reset_index()
```

```python
#analysing the statement - Women watch more Sci-Fi than men

# Female: ~10%
# Male: ~14%

# Women watch more Sci-Fi than men - FALSE

df_scifi = final_df['Sci-Fi'].div(final_df['count'])
df_scifi.reset_index()
```

```python
#analysing the statement - Men watch more Romance than women

# Female: ~23%
# Male: ~18%

# Men watch more Romance than women - FALSE

df_romance = final_df['Romance']/final_df['count']
df_romance.reset_index()
```
